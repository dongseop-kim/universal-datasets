### Example for public-TB

```python
%load_ext autoreload
%autoreload 2
%config Completer.use_jedi = False

from functools import partial
from matplotlib import pyplot as plt

from univdt.components import PublicTuberculosis
# print(PublicTuberculosis.TASK)
```

## Shenzhen

```python
root_dir = '' # TODO: set root dir
shenzhen = partial(PublicTuberculosis, root_dir=root_dir, transform=None, dataset='shenzhen')

shenzhen_train = shenzhen(split='train')
print(len(shenzhen_train))
assert len(shenzhen_train) == 530

shenzhen_val = shenzhen(split='val')
print(len(shenzhen_val))
assert len(shenzhen_val) == 66

shenzhen_test = shenzhen(split='test')
print(len(shenzhen_test))
assert len(shenzhen_test) == 66

shenzhen_test = shenzhen(split='trainval')
print(len(shenzhen_test))
assert len(shenzhen_test) == 596
```

```python
data= shenzhen_train.load_data(0)
image = data['image']
label = data['label']
print(image.shape, label) # 0 : normal, 1: TB, 2: Latent TB
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')

data= shenzhen_train.load_data(20)
image = data['image']
label = data['label']
print(image.shape, label) # 0 : normal, 1: TB, 2: Latent TB
plt.subplot(1,3,2)
plt.imshow(image, cmap='gray')

data= shenzhen_train.load_data(50)
image = data['image']
label = data['label']
print(image.shape, label) # 0 : normal, 1: TB, 2: Latent TB
plt.subplot(1,3,3)
plt.imshow(image, cmap='gray')
```
